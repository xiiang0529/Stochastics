```python
import math
import numpy as np
import numpy.random as npr
from pylab import plt, mpl
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family']='serif'
%config InlineBackend.figure_format='svg'
```

Random Numbers

```python
npr.seed(100)  # Set seed for random number generation to 100
np.set_printoptions(precision=4)  # Set the precision for printing to 4 decimal places
```

```python
npr.rand(10)  # Generate 10 random numbers
```

```python
npr.rand(5,5)  # Generate a 5x5 array of random numbers
```

```python
a=5  # Define variables a and b
b=10
npr.rand(10) * (b-a) + a  # Generate 10 random numbers between a and b
```

```python
npr.rand(5, 5) * (b-a) + a  # Generate a 5x5 array of random numbers between a and b
```

```python
sample_size = 500  # Define sample size
rn1 = npr.rand(sample_size, 3)  # Generate an array of random numbers with shape
rn2 = npr.randint(0, 10, sample_size)  # Generate an array of random integers between 0 and 10
rn3 = npr.sample(size=sample_size)  # Generate an array of random numbers
a = [0,25, 50, 75, 100]  # Define an array a
rn4 = npr.choice(a,size=sample_size)  # Generate an array of random numbers by sampling from array
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))  # Create a figure with 2x2 subplots
ax1.hist(rn1, bins=25, stacked=True) # Plot a histogram of rn1 on ax1
ax1.set_title('rand')
ax1.set_ylabel('frequency')
ax2.hist(rn2, bins=25)  # Plot a histogram of rn2 on ax2
ax2.set_title('randint')
ax3.hist(rn3, bins=25)  # Plot a histogram of rn3 on ax3
ax3.set_ylabel('frequency')
ax4.hist(rn4, bins=25)  # Plot a histogram of rn3 on ax4
ax4.set_title('chioce');
```
